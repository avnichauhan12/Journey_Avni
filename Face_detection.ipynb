```python
from mtcnn import MTCNN
from google.colab.patches import cv2_imshow
import cv2
detector=MTCNN()
img=cv2.imread('/single-person.jpg',cv2.IMREAD_UNCHANGED)
output= detector.detect_faces(img)
print(output)
x,y,width,height=output[0]['box']

cv2.rectangle(img,pt1=(x,y),pt2=(x+width,y+height),color=(255,0,0),thickness=3)
left_eyeX,left_eyeY = output[0]['keypoints']['left_eye']
cv2.circle(img,center=(left_eyeX,left_eyeY),color=(255,0,255),thickness=1,radius=5)
right_eyeX,right_eyeY = output[0]['keypoints']['right_eye']
cv2.circle(img,center=(right_eyeX,right_eyeY),color=(255,255,255),thickness=1,radius=5)
noseX,noseY = output[0]['keypoints']['nose']
cv2.circle(img,center=(noseX,noseY),color=(255,0,255),thickness=1,radius=5)



cv2_imshow(img)

```
